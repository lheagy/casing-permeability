## set up the simulations

```python
src_list = []
for loc in src_z:
    src_locations = np.array([
        [0, 0, loc+src_dz/2.],
        [0, 0, loc-src_dz/2.]
    ])
    src = tdem.sources.LineCurrent([], location=src_locations)
    src_list.append(src)
```

```python
locations = src.location
```

```python
for i in range(locations.shape[0] - 1):
    print(np.nonzero(np.abs(locations[i, :] - locations[i + 1, :]))[0])
```

```python
survey = tdem.Survey(src_list)
sim = tdem.simulation.Simulation3DMagneticField(
    mesh=mesh, sigmaMap=maps.IdentityMap(mesh),
    survey=survey, solver=Pardiso, time_steps=time_steps, 
#     verbose=True
)
```

```python
def load_fields(name, conductivity, permeability):
    solution = np.load(f"{directory}/fields_{name}.npy")
    sim.model = conductivity
    sim.mu = permeability
    f = sim.fieldsPair(sim)
    f[:, sim._fieldType + "Solution", :] = solution
    return f
```

```python
fields = {}

if run_simulations: 
    mesh.save(f"{directory}/mesh.json")
    np.save(f"{directory}/times", sim.times)
    
    fields["halfspace"] = sim.fields(halfspace_conductivity)
    np.save(f"{directory}/fields_halfspace", fields["halfspace"][:, f"{sim._fieldType}Solution", :])
else:
    fields["halfspace"] = load_fields("halfspace", halfspace_conductivity, mu_0*np.ones(mesh.n_cells))
```

```python
if run_simulations:
    for mur in casing_permeability:
        t = time.time()
        key = f"casing_{mur}"
        print(f"starting {key}")

        sim.mu = mu_0 * casing_permeability[mur]
        print(f"     check max permeability: {sim.mu.max()/mu_0}")

        fields[key] = sim.fields(casing_conductivity)
        np.save(f"{directory}/fields_{key}", fields[key][:, f"{sim._fieldType}Solution", :])
        print(f" ... done. {time.time()-t:1.2e} s")
else: 
    for mur in casing_permeability:
        key = f"casing_{mur}"
        fields[key] = load_fields(key, casing_conductivity, casing_permeability[mur])
```

```python
def plot_pos_neg_log(x, y, ax, plot_type="semilogy", color=None, label=None):
    if plot_type == "plot":
        return ax.plot(x, y, color=color, label=label)
    else:
        inds_pos = (y > 0)
        
        inds_switch = np.where(np.diff(inds_pos))[0].astype(int)+1
        starting_inds = np.hstack([np.r_[0], inds_switch])
        ending_inds = np.hstack([inds_switch, np.r_[len(inds_pos)]])
            
        for start, end in zip(starting_inds, ending_inds):
            xvals = x[start:end]
            yvals = y[start:end]
            if np.all(yvals <= 0):
                getattr(ax, plot_type)(xvals, np.abs(yvals), "--", color=color)
            else: 
                getattr(ax, plot_type)(xvals, yvals, "-", color=color, label=label)
                label=None
        return ax
```
